"""

"""

```python
#  Use the .plot extension from Pandas dataframe to make a scatterplot of the df features.
df.plot(kind="scatter", x="sepal_length", y="sepal_width")
```

```python
#  lets use the seaborn library to make a similar plot
#  A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x="sepal_length", y="sepal_width", data=df)
```

```python
# Use seaborn's FacetGrid to color the scatterplot by class
sns.FacetGrid(df, hue="class") \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()
```

```python
#  We can look at an individual feature in seaborn using a boxplot

# This plot shows that Iris-setosa can be separated from the other two species by a 
# straight line drawn horizontally from a petal length value of about 2.5
plt.figure(figsize=(20, 10))
sns.boxplot(x="class", y="petal_length", data=df)
```

```python
# Let's extend this plot by adding a layer of individual points on top of it through Seaborn's striplot
# We need to use jitter=True so that all the points don't fall in 
# single vertical line above the class (species)

# Saving the resulting axes as ax each time causes the resulting plot 
# to be shown on top of the previous axes
plt.figure(figsize=(20, 10))
ax = sns.boxplot(x="class", y="petal_length", data=df)
ax = sns.stripplot(x="class", y="petal_length", data=df, jitter=True, edgecolor="gray")
```

```python
#  A violin plot combines the benefits of the previous two plots and simplifies them
#  Denser regions of the data are fatter, and sparser thiner in a violin plot
plt.figure(figsize=(20, 10))
sns.violinplot(x="class", y="petal_length", data=df, size=6)
```

```python
#  Another plot useful for looking at univariate relations is the kdeplot, which creates 
# and visualizes a kernel density estimate of the underlying feature.

#  We again see that Iris-setosa is linearly separable from the other two class of species,
#  but Iris-versicolor and Iris-virginica overlap in petal-length.
sns.FacetGrid(df, hue="class", size=6) \
   .map(sns.kdeplot, "petal_length") \
   .add_legend()
```

```python
#  Now the below is the seaborn pairplot, which shows the bivariate relation between 
# each pair of features

#  From the pairplot, we will see that the Iris-setosa class species is separated from the other
#  two across all feature combinations
sns.pairplot(df, hue="class", size=3)
```

```python
#  Below Seaborn helps to computes and plots the regression for each atribute, all class combined

sns.pairplot(df, kind="reg", size=3)
```

```python
# We could also compute one regression for each class of species
sns.pairplot(df, hue="class", kind="reg")
```

```python
# Now we will see how features are correlated to each other using heatmap in seaborn library. 
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True, cmap="YlGnBu")
plt.plot()
```
