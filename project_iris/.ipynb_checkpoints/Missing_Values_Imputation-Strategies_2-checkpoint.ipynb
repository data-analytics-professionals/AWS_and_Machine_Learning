"""
Working with Missing Data in Pandas
Missing Data can occur when no information is provided for one or more items or for a whole unit. 
Missing Data is a very big problem in real life scenario. 
Missing Data can also refer to as NA(Not Available) values in pandas. 

In DataFrame sometimes many datasets simply arrive with missing data, either because 
it exists and was not collected or it never existed. 
For Example, Suppose different user being surveyed may choose not to share their income, 
some user may choose not to share the address 
in this way many datasets went missing.

In Pandas missing data is represented by two value:

None: None is a Python singleton object that is often used for missing data in Python code.
NaN : NaN (an acronym for Not a Number), is a special floating-point value recognized by all systems 
that use the standard IEEE floating-point representation

Pandas treat None and NaN as essentially interchangeable for indicating missing or null values. 
To facilitate this convention, there are several useful functions for detecting, removing, and 
replacing null values in Pandas DataFrame. Please stay connected to watch my separate lecture 
in this series to learn more about on how to deal with missing values. 

"""

```python
# innull() function detect missing values and when used with sum() function, it becomes more useful.
df.isnull().sum()
```

```python
# Fill NA/NaN values using the specified method.
df = df.fillna(df.mean())
```

```python
# check if NaN are replaced by mean of the column.
df.isnull().sum()
```
